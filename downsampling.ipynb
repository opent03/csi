```python
import numpy as np
```

```python
X_train, X_test = np.load('X_train.npy'), np.load('X_test.npy')
```

```python
X_train.shape #sanity check
```

```python
def downsample(array):
    new_array = []
    for i in range(len(array)):
        sample = array[i]
        t = int(len(sample)/2)
        sample = sample[(t-1000):(t+1000),]
        sample = sample[::20]
        new_array.append(sample)
    return np.array(new_array)
```

```python
X_train, X_test = downsample(X_train), downsample(X_test)
```

```python
X_train, X_test = np.array(X_train), np.array(X_test)
```
