```python
def moving_average(data, window_size):
    window = np.ones(int(window_size))/float(window_size)
    return np.convolve(data, window, 'same')

def visualize(path1):
    #data import
    data = pd.read_csv(path1, header=None).values
    amp = data[:,1:91]

    #plt
    fig = plt.figure(figsize = (18,10))
    ax1 = plt.subplot(311)
    plt.imshow(amp[:,0:29].T,interpolation = "nearest", aspect = "auto", cmap="jet")
    ax1.set_title("Antenna1 Amplitude")
    plt.colorbar()

    ax2 = plt.subplot(312)
    plt.imshow(amp[:,30:59].T,interpolation = "nearest", aspect = "auto", cmap="jet")
    ax2.set_title("Antenna2 Amplitude")
    plt.colorbar()

    ax3 = plt.subplot(313)
    plt.imshow(amp[:,60:89].T,interpolation = "nearest", aspect = "auto", cmap="jet")
    ax3.set_title("Antenna3 Amplitude")
    plt.colorbar()
    plt.show()
    
    # Initializing valiables
    constant_offset = np.empty_like(amp)
    filtered_data = np.empty_like(amp)

    # Calculating the constant offset (moving average 4 seconds)
    for i in range(1, len(amp[0])):
        constant_offset[:,i] = moving_average(amp[:,i], 4000)

    # Calculating the filtered data (substract the constant offset)
    filtered_data = amp - constant_offset

    # Smoothing (moving average 0.01 seconds)
    for i in range(1, len(amp[0])):
        filtered_data[:,i] = moving_average(filtered_data[:,i], 10)
    # Calculate correlation matrix (90 * 90 dim)
    cov_mat2 = np.cov(filtered_data.T)
    # Calculate eig_val & eig_vec
    eig_val2, eig_vec2 = np.linalg.eig(cov_mat2)
    # Sort the eig_val & eig_vec
    idx = eig_val2.argsort()[::-1]
    eig_val2 = eig_val2[idx]
    eig_vec2 = eig_vec2[:,idx]
    # Calculate H * eig_vec
    pca_data2 = filtered_data.dot(eig_vec2)
    
    xmin = 0
    xmax = 200000  
    # plt
    fig3 = plt.figure(figsize = (18,20))

    ax1 = plt.subplot(611)
    plt.plot(pca_data2[xmin:xmax,0])
    #plt.plot(pca_data2[2500:17500,0])
    ax1.set_title("PCA 1st component")

    ax2 = plt.subplot(612)
    plt.plot(pca_data2[xmin:xmax,1])
    #plt.plot(pca_data2[2500:17500,1])
    ax2.set_title("PCA 2nd component")

    ax3 = plt.subplot(613)
    plt.plot(pca_data2[xmin:xmax,2])
    #plt.plot(pca_data2[2500:17500,2])
    ax3.set_title("PCA 3rd component")

    ax4 = plt.subplot(614)
    plt.plot(pca_data2[xmin:xmax,3])
    #plt.plot(pca_data2[2500:17500,3])
    ax4.set_title("PCA 4th component")

    ax5 = plt.subplot(615)
    plt.plot(pca_data2[xmin:xmax,4])
    #plt.plot(pca_data2[2500:17500,4])
    ax5.set_title("PCA 5th component")

    ax6 = plt.subplot(616)
    plt.plot(pca_data2[xmin:xmax,5])
    #plt.plot(pca_data2[2500:17500,5])
    ax6.set_title("PCA 6th component")

    plt.show()
    
    plt.figure(figsize = (18,30))
    # Spectrogram(STFT)
    plt.subplot(611)
    Pxx, freqs, bins, im = plt.specgram(pca_data2[:,0], NFFT=256, Fs=1000, noverlap=1, cmap="jet", vmin=-100,vmax=20)
    plt.xlabel("Time[s]")
    plt.ylabel("Frequency [Hz]")
    plt.title("Spectrogram(STFT) Component 1")
    plt.colorbar(im)
    #plt.xlim(0,20)
    plt.ylim(0,80)

    plt.subplot(612)
    Pxx, freqs, bins, im = plt.specgram(pca_data2[:,1], NFFT=256, Fs=1000, noverlap=1, cmap="jet", vmin=-100,vmax=20)
    plt.xlabel("Time[s]")
    plt.ylabel("Frequency [Hz]")
    plt.title("Spectrogram(STFT) Component 2nd")
    plt.colorbar(im)
    #plt.xlim(0,20)
    plt.ylim(0,80)

    plt.subplot(613)
    Pxx, freqs, bins, im = plt.specgram(pca_data2[:,2], NFFT=256, Fs=1000, noverlap=1, cmap="jet", vmin=-100,vmax=20)
    plt.xlabel("Time[s]")
    plt.ylabel("Frequency [Hz]")
    plt.title("Spectrogram(STFT) Component 3rd")
    plt.colorbar(im)
    #plt.xlim(0,20)
    plt.ylim(0,80)

    plt.subplot(614)
    Pxx, freqs, bins, im = plt.specgram(pca_data2[:,3], NFFT=256, Fs=1000, noverlap=1, cmap="jet", vmin=-100,vmax=20)
    plt.xlabel("Time[s]")
    plt.ylabel("Frequency [Hz]")
    plt.title("Spectrogram(STFT) Component 4rd")
    plt.colorbar(im)
    #plt.xlim(0,200)
    plt.ylim(0,80)

    plt.subplot(615)
    Pxx5, freqs5, bins5, im = plt.specgram(pca_data2[:,4], NFFT=256, Fs=1000, noverlap=1, cmap="jet", vmin=-100,vmax=15)
    plt.xlabel("Time[s]")
    plt.ylabel("Frequency [Hz]")
    plt.title("Spectrogram(STFT) Component 5th")
    plt.colorbar(im)
    #plt.xlim(0,200)
    plt.ylim(15,80)
   
    
    plt.subplot(616)
    Pxx, freqs, bins, im = plt.specgram(pca_data2[:,5], NFFT=256, Fs=1000, noverlap=1, cmap="jet", vmin=-100,vmax=15)
    plt.xlabel("Time[s]")
    plt.ylabel("Frequency [Hz]")
    plt.title("Spectrogram(STFT) Component 6th")
    plt.colorbar(im)
    #plt.xlim(0,200)
    plt.ylim(15,80)
    
    plt.show()
    
    plt.figure(figsize = (18,10))
    ax = plt.subplot(111)
#    ax.magnitude_spectrum(pca_data2[:,0], Fs=1000, scale='dB', color='C1')
    ax.magnitude_spectrum(pca_data2[5000:7500,0], Fs=1000, color='C1')
    plt.xlim(0,50)
    plt.ylim(0,10)
    plt.show()
    
    return Pxx5, freqs5, bins5
```

```python
def visualize(path1, load_data=True, data=None):
    #data import
    if load_data:
        data = pd.read_csv(path1, header=None).values
    amp = data[:,1:91]

    #plt
    fig = plt.figure(figsize = (18,10))
    ax1 = plt.subplot(311)
    plt.imshow(amp[:,0:29].T,interpolation = "nearest", aspect = "auto", cmap="jet")
    ax1.set_title("Antenna1 Amplitude")
    plt.colorbar()

    ax2 = plt.subplot(312)
    plt.imshow(amp[:,30:59].T,interpolation = "nearest", aspect = "auto", cmap="jet")
    ax2.set_title("Antenna2 Amplitude")
    plt.colorbar()

    ax3 = plt.subplot(313)
    plt.imshow(amp[:,60:89].T,interpolation = "nearest", aspect = "auto", cmap="jet")
    ax3.set_title("Antenna3 Amplitude")
    plt.colorbar()
    plt.show()
    
    # Initializing valiables
    constant_offset = np.empty_like(amp)
    filtered_data = np.empty_like(amp)

    # Calculating the constant offset (moving average 4 seconds)
    for i in range(1, len(amp[0])):
        constant_offset[:,i] = moving_average(amp[:,i], 4000)

    # Calculating the filtered data (substract the constant offset)
    filtered_data = amp - constant_offset

    # Smoothing (moving average 0.01 seconds)
    for i in range(1, len(amp[0])):
        filtered_data[:,i] = moving_average(filtered_data[:,i], 10)
    # Calculate correlation matrix (90 * 90 dim)
    cov_mat2 = np.cov(filtered_data.T)
    # Calculate eig_val & eig_vec
    eig_val2, eig_vec2 = np.linalg.eig(cov_mat2)
    # Sort the eig_val & eig_vec
    idx = eig_val2.argsort()[::-1]
    eig_val2 = eig_val2[idx]
    eig_vec2 = eig_vec2[:,idx]
    # Calculate H * eig_vec
    pca_data2 = filtered_data.dot(eig_vec2)

    xmin = 0
    xmax = 200000
    
    # PCA Plots
    fig3 = plt.figure(figsize = (18,20))
    ax_pca = []
    k = 611
    for i in range(6):
        ax = plt.subplot(k + i)
        plt.plot(pca_data2[xmin:xmax,i])
        #plt.plot(pca_data2[2500:17500,0])
        ax.set_title("PCA {} component".format(i+1))
        ax_pca.append(ax)

    plt.show()
    
    # EDIT THIS
    k = 611
    
    plt.figure(figsize = (18,30))
    Pxx_arr, freqs_arr, bins_arr, im_arr = [], [], [], []
    
    # Spectrogram(STFT)
    for i in range(6):
        plt.subplot(k + i)
        Pxx, freqs, bins, im = plt.specgram(pca_data2[:,i], NFFT=256,
                                            Fs=1000, noverlap=1, cmap="jet", vmin=-100,vmax=20)
        plt.xlabel("Time[s]")
        plt.ylabel("Frequency [Hz]")
        plt.title("Spectrogram(STFT) Component {}".format(i+1))
        plt.colorbar(im)
        #plt.xlim(0,20)
        plt.ylim(0,80)
        
        Pxx_arr.append(Pxx)
        freqs_arr.append(freqs)
        bins_arr.append(bins)
        im_arr.append(im)
        
        
    plt.show()
    
    plt.figure(figsize = (18,10))
    ax = plt.subplot(111)
#    ax.magnitude_spectrum(pca_data2[:,0], Fs=1000, scale='dB', color='C1')
    ax.magnitude_spectrum(pca_data2[5000:7500,0], Fs=1000, color='C1')
    plt.xlim(0,50)
    plt.ylim(0,10)
    plt.show()
    
    return Pxx_arr, freqs_arr, bins_arr, pca_data2
```

```python
##### main freqs components from ~10 Hz to 27 Hz
##### GET CSI COMPONENTS
bins_mvaPeaks=bins5[mvaPeaks]
print(bins_mvaPeaks)
bins_mvaPeaks = list(map(lambda x: int(x * 1000), bins_mvaPeaks))
print(bins_mvaPeaks)
def get_avg_split(lst):
    import statistics
    diff = []
    for i in range(len(lst)):
        if i == 0:
            continue
        diff.append(lst[i] - lst[i-1])
    return round(statistics.mean(diff))
avg_split = get_avg_split(bins_mvaPeaks)
CSI = pd.read_csv("Dataset/cross_loc1_consolidated.csv", header=None).values # read the same thing into a np array
CSI.shape
```

```python
print(CSI[bins_mvaPeaks[1]:bins_mvaPeaks[2]].shape)
```

```python
print(CSI.shape) # avg split determines the y dimension in the tensor (x, y, z) in which to feed into the neural network
def get_CSI_partition(CSI, peaks, split):
    partition = np.empty(shape=(len(peaks), split*2, CSI.shape[1]))
    print(partition.shape)
    for i in range(len(peaks)): # we assume that data collection started early enough and ended soon enough,
        # so we won't make an exception for first and last peaks in case the window goes beyond the first and 
        # last indices. Otherwise, just disregard first and last peak, losing data
        partition[i] = CSI[(peaks[i]-split):(peaks[i]+split)]
    return partition
avg_split = 4000
CSI_part = get_CSI_partition(CSI, bins_mvaPeaks, int(avg_split/2))
CSI_part.shape
### CSI_part is a rank 3 tensor of dimensions (batches, time, CSI_information)
### time is determined by the average difference between the peak, halved
### It is best to use the same time interval (window size) for all processing, so the dimensions match up 
### with the dimensions of the weights in the ResNets
```
